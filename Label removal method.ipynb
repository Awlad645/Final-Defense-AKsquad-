```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
import cv2
import seaborn as sns
import os
import glob
```

```python
path = "/content/Normal (1).jpg"
img= cv2.imread(path)#BGR
cpy_img = img.copy()
lower = np.array([50,50,50])
higher =np.array([250,250,250])
mask = cv2.inRange(img,lower,higher)
cont,_=cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
cont_img=cv2.drawContours(img,cont,-1,255,3)
c=max(cont,key=cv2.contourArea)
x,y,w,h=cv2.boundingRect(c)
cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),7)
crroped_image = cpy_img[y:y+h,x:x+w]

plt.figure(figsize=(20,4))
plt.subplot(1,4,1),plt.imshow(cpy_img),plt.xlabel("Orginal Image")
plt.subplot(1,4,2),plt.imshow(mask,'gray'),plt.xlabel("Mask Image")

cv2.imwrite("mask.png", mask)

plt.subplot(1,4,3),plt.imshow(img),plt.xlabel("Contour Image")
cv2.imwrite("contor.png", img)
plt.subplot(1,4,4),plt.imshow(crroped_image),plt.xlabel("Crroped Image")
cv2.imwrite("crroped.png", crroped_image)
plt.show()
```
